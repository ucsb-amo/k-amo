```python
# wavelengths = np.linspace(1050.,1140.,1000) * 1.e-9
wavelengths = np.linspace(1210.,1280.,500) * 1.e-9
alpha_4s, _, _, _ = cp.compute_fine_structure_polarizability(4,0,1/2,wavelength_m=wavelengths,portal_data=portal_data)
alpha_4p, _, _, _ = cp.compute_fine_structure_polarizability(4,1,1/2,wavelength_m=wavelengths,portal_data=portal_data)
```

```python
plt.figure()
yylim = [-20000,20000]
msize = 2
plt.plot(wavelengths*1.e9,alpha_4s,'.',markersize=msize)
plt.plot(wavelengths*1.e9,alpha_4p,'.',markersize=msize)
plt.xlabel("wavelength (nm)")
plt.ylabel("polarizability")
plt.legend(['4s','4p1/2'])
plt.ylim(yylim)
plt.vlines(1227.67,ymin=yylim[0],ymax=yylim[1],linestyles='--',colors='k')
# plt.xlim([1220,1228])
plt.show()
```
